```python
word = input("Enter a word: ")

if word == word[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
num = int(input("Enter a number: "))
original = num
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10
if original == reverse:
    print("The number is a palindrome.")
else:
    print("The number is not a palindrome.")
```

```python
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")
```

```python

num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):
    fact = fact * i
print(f"The factorial of {num} is {fact}.")
```

```python
def table(num):
    for i in range(1, 11):
        print(num, "X", i, "=", num * i)

num = int(input("Enter a number: "))
table(num)
```
